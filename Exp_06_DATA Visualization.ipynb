# Line Chart

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

# Bar Chart

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x = np.arange(1, 11)                 # x values (1 to 10)
y = np.random.randint(1, 100, 10)    # random integers for y

# Create bar chart
plt.bar(x, y, color='skyblue', edgecolor='black')

# Add labels and title
plt.title("Random Bar Chart")
plt.xlabel("X Axis")
plt.ylabel("Y Axis (Random Values)")
```
